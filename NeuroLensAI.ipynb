```python
# BUILDING DATASET LOADER
import os
import numpy as np
import nibabel as nib
from torch.utils.data import Dataset

class MRISliceDataset(Dataset):
    def __init__(self, data_dir, labels_df, sequence="t1c", transform=None):
        """
        data_dir: folder containing one subfolder per patient (BraTS-PED-XXXXX-000)
        labels_df: dataframe with columns ['BraTS-SubjectID', 'subtype_label']
        """
        self.transform = transform
        self.class_names = sorted(labels_df["subtype_label"].unique())
        self.class_to_idx = {c: i for i, c in enumerate(self.class_names)}

        self.samples = []
        for _, row in labels_df.iterrows():
            patient_id = row["BraTS-SubjectID"]
            patient_dir = os.path.join(data_dir, patient_id)
            scan_path = self._find_sequence_file(patient_dir, sequence)
            if scan_path:
                label_idx = self.class_to_idx[row["subtype_label"]]
                self.samples.append((scan_path, label_idx))

        print(f"Built dataset: {len(self.samples)} patients found out of {len(labels_df)} labeled")

    @staticmethod
    def _find_sequence_file(patient_dir, sequence):
        if not os.path.isdir(patient_dir):
            return None
        for fname in os.listdir(patient_dir):
            if sequence.lower() in fname.lower() and fname.endswith((".nii", ".nii.gz")):
                return os.path.join(patient_dir, fname)
        return None

    @staticmethod
    def _find_seg_file(patient_dir):
        if not os.path.isdir(patient_dir):
            return None
        for fname in os.listdir(patient_dir):
            if "seg" in fname.lower() and fname.endswith((".nii", ".nii.gz")):
                return os.path.join(patient_dir, fname)
        return None

    @staticmethod
    def _find_best_slice(seg_volume):
        # priority order: prefer enhancing tumor (4) if present, since it's the
        # most tumor-specific tissue; fall back to necrotic/non-enhancing core (1)
        # if no enhancement exists for this patient (common in DMG/DIPG, which
        # often doesn't enhance); fall back further to any tumor-related tissue
        # (including edema, label 2) only if neither core label is present at all
        for label_value in [4, 1]:
            core_tumor_mask = (seg_volume == label_value)
            tumor_voxel_counts = core_tumor_mask.sum(axis=(0, 1))
            best_slice_idx = tumor_voxel_counts.argmax()
            if tumor_voxel_counts[best_slice_idx] > 0:
                return best_slice_idx

        # neither label 4 nor label 1 present anywhere — fall back to edema/any tumor tissue
        any_tumor_mask = (seg_volume > 0)
        fallback_counts = any_tumor_mask.sum(axis=(0, 1))
        return fallback_counts.argmax()

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
      filepath, label = self.samples[idx]
      volume = nib.load(filepath).get_fdata()

      patient_dir = os.path.dirname(filepath)
      seg_path = self._find_seg_file(patient_dir)

      if seg_path:
          seg_volume = nib.load(seg_path).get_fdata()
          best_slice = self._find_best_slice(seg_volume)
      else:
          best_slice = volume.shape[2] // 2

      # normalize using the WHOLE volume's intensity range, not just this slice —
      # keeps contrast consistent across different slices/patients
      volume = np.nan_to_num(volume)
      volume_norm = (volume - volume.min()) / (volume.max() - volume.min() + 1e-8)
      slice_2d = volume_norm[:, :, best_slice]

      slice_2d = (slice_2d * 255).astype(np.uint8)
      slice_3ch = np.stack([slice_2d] * 3, axis=-1)

      if self.transform:
          slice_3ch = self.transform(slice_3ch)

      return slice_3ch, label
```

```python
# BUILDING THE MODEL
import torch
import torch.nn as nn
from torchvision import models

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def build_model(num_classes, freeze_backbone=True):
    model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)

    if freeze_backbone:
        for param in model.parameters():
            param.requires_grad = False

    model.fc = nn.Sequential(
        nn.Dropout(0.3),
        nn.Linear(model.fc.in_features, num_classes),
    )
    return model.to(DEVICE)

model = build_model(num_classes=len(dataset.class_names))
```

```python
# TRAIN THE MODEL
import numpy as np
from torch.utils.data import DataLoader

BATCH_SIZE = 16
EPOCHS = 25
LR = 1e-4

train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_set, batch_size=BATCH_SIZE)
test_loader = DataLoader(test_set, batch_size=BATCH_SIZE)

train_labels = [dataset.samples[i][1] for i in train_idx]
class_counts = np.bincount(train_labels, minlength=len(dataset.class_names))
class_weights = torch.tensor(
    [1.0 / c if c > 0 else 0.0 for c in class_counts], dtype=torch.float32
).to(DEVICE)
print(f"Class weights: {dict(zip(dataset.class_names, class_weights.tolist()))}")

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.Adam(
    filter(lambda p: p.requires_grad, model.parameters()), lr=LR
)

for epoch in range(EPOCHS):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{EPOCHS} — training loss: {running_loss/len(train_loader):.4f}")
```

```python
# CHECK IF IT WORKS
from sklearn.metrics import classification_report, confusion_matrix

def evaluate(model, loader):
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images)
            preds = outputs.argmax(dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    print(classification_report(all_labels, all_preds, target_names=dataset.class_names))
    print("Confusion matrix:\n", confusion_matrix(all_labels, all_preds))

evaluate(model, test_loader)
```

```python
def get_predictions_with_ids(model, loader, dataset, indices_used):
    model.eval()
    all_preds, all_labels, all_patient_ids = [], [], []

    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images)
            preds = outputs.argmax(dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # match predictions back to patient IDs, in the same order the loader served them
    # NOTE: only works reliably if shuffle=False on this loader (test_loader already is)
    patient_ids = [dataset.samples[i][0].split('/')[-2] for i in indices_used]

    return list(zip(patient_ids, all_labels, all_preds))

results = get_predictions_with_ids(model, test_loader, dataset, test_idx)
```

```python
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget

target_layers = [model.layer4[-1]]  # the last block within layer4

cam = GradCAM(model=model, target_layers=target_layers)

targets = [ClassifierOutputTarget(predicted_class)]  # explain THIS specific prediction

grayscale_cam = cam(input_tensor=input_tensor, targets=targets)
grayscale_cam = grayscale_cam[0, :]  # remove batch dimension — now a 2D heatmap
```

```python
# convert the normalized tensor back to a viewable 0-1 range image
img_for_display = image_tensor.permute(1, 2, 0).cpu().numpy()  # [C,H,W] -> [H,W,C]

mean = np.array([0.485, 0.456, 0.406])
std = np.array([0.229, 0.224, 0.225])
img_for_display = img_for_display * std + mean  # undo the normalization
img_for_display = np.clip(img_for_display, 0, 1)  # keep values in valid 0-1 range
```

```python
from pytorch_grad_cam.utils.image import show_cam_on_image

visualization = show_cam_on_image(img_for_display, grayscale_cam, use_rgb=True)
```
